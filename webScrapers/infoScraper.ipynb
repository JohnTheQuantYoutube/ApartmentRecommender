## get_data

This first function is going to go through all of the placards on the apartments.com page for Denver, Colorado. We are going to save all of the information, especially apartment names which we can then use to scrape the reviews.

```python
# number of pages
num_pages = 28
```

```python
# get_data function
def get_data(page_number):
  # idk...just took this line from the DataCamp tutorial
  headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

  r = requests.get('https://www.apartments.com/denver-co/' + str(page_number) + '/', 
                   headers=headers)
  content = r.content
  soup = BeautifulSoup(content)
  
  names = []
  urls = []
  addresses = []
  for prop in soup.findAll('div', attrs={'class':'property-information'}):
    names.append(prop.find('span', attrs={'class':'js-placardTitle title'}).text)
    urls.append(prop.find('a', attrs={'class':'property-link'})['href'])
    addresses.append(prop.find('div', attrs={'class':'property-address js-url'}).text)
  df = pd.DataFrame([names, urls, addresses]).T
  df.columns = ['names', 'urls', 'addresses']
  return df
```

```python
whole_df = pd.DataFrame(columns=['names', 'urls', 'addresses'])
for page in range(1, num_pages+1):
  print(f'Scraping page {page}...')
  whole_df = pd.concat([whole_df, get_data(page)])
print('Done!')
```
